## Transformation des données

###  Extension de chaque dictionnaire de la liste data_rapport_naissances par les colonnes suivantes :
-Extraction des valeurs de la clé 'contact' :
 'contact_uuid',  'contact_name', 'contact_urn'
 
-Extraction des valeurs de la clé 'values' :
 'confirmation_finale', 'depassement_delai_rapportage'

```python
for a in  range(0,len(data_rapport_naissances) - 1):
    
    #Extraction des valeurs de la clé 'contact'
    contact_uuid = data_rapport_naissances[a].get('contact').get('uuid')
    contact_urn = data_rapport_naissances[a].get('contact').get('urn')
    contact_name =data_rapport_naissances[a].get('contact').get('name')
    
    #Extraction des valeurs de la clé 'values'
    if 'confirmation_finale' in data_rapport_naissances[a].get('values'):
        confirmation_finale = data_rapport_naissances[a].get('values').get('confirmation_finale').get('value')    
    if 'depassement_delai_rapportage' in data_rapport_naissances[a].get('values'):
        depassement_delai_rapportage = data_rapport_naissances[a].get('values').get('depassement_delai_rapportage').get('value')
        data_rapport_naissances[a]['depassement_delai_rapportage'] = depassement_delai_rapportage        
 
    data_rapport_naissances[a].update({'contact_uuid' : contact_uuid,
                                       'contact_urn' : contact_urn,
                                       'contact_name' : contact_name,
                                       'confirmation_finale': confirmation_finale})        
 
```

### Conversion de la liste 'data_rapport_naissances' en dataFrame

```python
import pandas as pd
data_rapport_naissances = pd.DataFrame(data_rapport_naissances)
data_rapport_naissances
```

### Division de la colonne 'created_on' en deux colonnes différentes :
- 'create_on_date'
- 'create_on_heure'

```python

data_rapport_naissances=data_rapport_naissances.join(data_rapport_naissances['created_on'].str.split('T', 1, expand=True).rename(columns={0:'created_on_date', 1:'created_on_heure'}))
```

#### Conversion de la colonne 'created_on_date' au format datatime

```python
data_rapport_naissances.created_on_date= pd.to_datetime(data_rapport_naissances.created_on_date) 
```

#### Creation des colonnes: 'created_on_year', 'created_on_month', 'created_on_day'

```python
data_rapport_naissances['created_on_year'] = data_rapport_naissances['created_on_date'].dt.year
data_rapport_naissances['created_on_month'] = data_rapport_naissances['created_on_date'].dt.month
data_rapport_naissances['created_on_day'] = data_rapport_naissances['created_on_date'].dt.day
```

### Suppression des colonnes qu'on ne vas pas utiliser

```python
data_rapport_naissances = data_rapport_naissances.drop(['uuid','flow', 'contact', 'start', 
                                    'path', 'values', 'exited_on', 'created_on', 'created_on_date',], axis=1)
```
